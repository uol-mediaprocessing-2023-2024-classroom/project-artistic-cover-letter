# YOLO

```python
import cv2 
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import PIL
import glob
import os
from PIL import Image
from random import randrange
from ultralytics import YOLO
```

```python
# load a pretrained model (recommended for training)
model = YOLO("yolov8m.pt")
```

```python
# Load an image
#img = ('test1.jpeg')
img = ('img_0116.jpg')

# Run batched inference on a image
results = model(img)
```

```python
# Process results list
for result in results:
    boxes = result.boxes  # Boxes object for bbox outputs
    print(boxes)
```

```python
# Get the height and width of the image
save = cv2.imread(img)
(h, w) = save.shape[:2]
```

# First try: problem with center and margin

```python
# Load image
im = Image.open(img)
fig, ax = plt.subplots()
ax.imshow(im)

# Initialization of centre list
centre = []

# For all result detected by YOLO 
for result in results:
    for i, boxe in enumerate(result.boxes.conf):
        # If YOLO is 85% sure of the object class
        if boxe.numpy() > 0.80:
            xyxy = result.boxes.xyxy.numpy()

            # Generates rectangles around objects
            rect = patches.Rectangle((xyxy[i][0], xyxy[i][1]), (xyxy[i][2]-xyxy[i][0]), (xyxy[i][3]-xyxy[i][1]), linewidth=1, edgecolor='r', facecolor='none')
            ax.add_patch(rect)

            # Get the center of the box
            centre.append([round((xyxy[i][2]+xyxy[i][0])/2), round((xyxy[i][3]+xyxy[i][1])/2)])

plt.show()
print('List of all red box centers: ' +str(centre))
```

```python
# Length and width of cropped image 
square = 150

# Crop image around red box center
for i, cen in enumerate(centre) :
    box = (cen[0]-square, cen[1]-square, cen[0]+square, cen[1]+square)
    cropImage = im.crop(box)
    cropImage.save('crop/firstTry'+str(i)+'.jpg')
```

```python
for filename in glob.glob('crop/firstTry*'):
    image = Image.open(filename)
    fig, ax = plt.subplots()
    ax.imshow(image)
    plt.axis('off')
    plt.show()
```

# Second try: problem with margin fixed (problem with center still here)

```python
# Load image
im = Image.open(img)
fig, ax = plt.subplots()
ax.imshow(im)

# Initialization of centre list
centre = []

# For all result detected by YOLO 
for result in results:
    for i, boxe in enumerate(result.boxes.conf):
        # If YOLO is 85% sure of the object class
        if boxe.numpy() > 0.80:
            xyxy = result.boxes.xyxy.numpy()

            # Generates rectangles around objects
            rect = patches.Rectangle((xyxy[i][0], xyxy[i][1]), (xyxy[i][2]-xyxy[i][0]), (xyxy[i][3]-xyxy[i][1]), linewidth=1, edgecolor='r', facecolor='none')
            ax.add_patch(rect)

            # Get the center of the box (x, y)
            centre.append([round((xyxy[i][2]+xyxy[i][0])/2), round((xyxy[i][3]+xyxy[i][1])/2)])

plt.show()
print('List of all red box centers: ' +str(centre))
```

```python
# Length and width of cropped image 
square = 150

for i, cen in enumerate(centre) :
    if (cen[0]-square) < 0:
        cen[0] = cen[0] + -1*(cen[0]-square)
    if (cen[0]+square) > w:
        cen[0] = cen[0] - -1*(w-(cen[0]+square))
    if (cen[1]-square) < 0:
        cen[1] = cen[1] + -1*(cen[1]-square)  
    if (cen[1]+square) > h:
        cen[1] = cen[1] - -1*(h-(cen[1]+square))

    box = (cen[0]-square, cen[1]-square, cen[0]+square, cen[1]+square)
    cropImage = im.crop(box)
    cropImage.save('crop/secondTry'+str(i)+'.jpg')
```

```python
for filename in glob.glob('crop/secondTry*'):
    image = Image.open(filename)
    fig, ax = plt.subplots()
    ax.imshow(image)
    plt.axis('off')
    plt.show()
```

# Third try: problem with margin and center fixed

```python
# Load image
im = Image.open(img)
fig, ax = plt.subplots()
ax.imshow(im)

# Initialization of centre list
centre = []

# For all result detected by YOLO 
for result in results:
    for i, boxe in enumerate(result.boxes.conf):
        # If YOLO is 85% sure of the object class
        if boxe.numpy() > 0.80:
            xyxy = result.boxes.xyxy.numpy()

            # Generates rectangles around objects
            rect = patches.Rectangle((xyxy[i][0], xyxy[i][1]), (xyxy[i][2]-xyxy[i][0]), (xyxy[i][3]-xyxy[i][1]), linewidth=1, edgecolor='r', facecolor='none')
            ax.add_patch(rect)

            # Get the center of the box
            centre.append([round((xyxy[i][2]+xyxy[i][0])/2), round(xyxy[i][1])])

plt.show()
print('List of all red box centers: ' +str(centre))
```

```python
# Length and width of cropped image 
square = 150

for i, cen in enumerate(centre) :
    if (cen[0]-square) < 0:
        cen[0] = cen[0] + -1*(cen[0]-square)
    if (cen[0]+square) > w:
        cen[0] = cen[0] - -1*(w-(cen[0]+square))
    if cen[1]+2*square > h:
        cen[1] = cen[1] + -1*(h-(cen[1]+2*square))
    box = (cen[0]-square, cen[1], cen[0]+square, cen[1]+2*square)
    cropImage = im.crop(box)
    cropImage.save('crop/thirdTry'+str(i)+'.jpg')
```

```python
for filename in glob.glob('crop/thirdTry*'):
    image = Image.open(filename)
    fig, ax = plt.subplots()
    ax.imshow(image)
    plt.axis('off')
    plt.show()
```
